### 벡터와 행렬 선언

```python
a = np.array([1,2,3])
print(a)
```

```python
print(np.ones((1,5))) # 행, 열 순으로 입력
print(np.ones((3,2)))
print(np.zeros((1,5)))
```

```python
# 주어진 범위 안에서 순차적으로 증가하는 리스트
# np.arange([start,] stop, [step,] dtype=None)

print(np.arange(10))
print(np.arange(3,7, dtype=np.float))
print(np.arange(3,10,2))
```

```python
# 행렬 선언
mat1 = np.array([[1,2,3],[4,5,6]])

# random 초기화
mat2 = np.random.randint(low=1, high=10, size=(3,2)) # 1<=x<=10으로 3*2 matrix 생성
mat3 = np.random.rand(3,2) # 정규분포를 따르는

print(mat1)
print(mat2)
print(mat3)
```

### np.reshape


```python
mat1 = np.random.rand(6,3)
print(mat1)
print(mat1.reshape(1,-1)) # -1 == all
print(mat1.reshape(-1,1))
# 사이즈가 맞지 않을 경우 err 출력
# tensor(3차원 이상 행렬) 형태로도 가능
print(mat1.reshape(3,2,3))
```

```python
# matrix/tensor 형태에서의 slicing
mat2 = np.arange(24).reshape(-1,4)
print(mat2)
print(mat2[:1,:3])
print(mat2[3, 0:2])
```

### 사칙연산


```python
x = np.array([[1,3,5],[7,9,11],[13,15,17]])
y = np.array([[2,4,6],[8,10,12],[14,16,18]])
```

```python
# 합
print(x.sum(axis=0))
print(x.sum(axis=1))
# 평균
print(x.mean(axis=0))
# 표준편차
print(x.std(axis=0))
```

```python
# transpose
print(x.T)
```

### 기타 연산

## Pandas

```python
import pandas as pd
```

### pandas series 만들기

```python
pd_series = pd.Series(index=['a','b','c','d','e'], data=[1,2,3,4,5])
pd_series
```

```python
# series의 기초 정보
print('차원:', pd_series.ndim)
print('형태: ', pd_series.shape)
print('총 원소의 수:', pd_series.size)

print('값:', pd_series.values)
print('인덱스:', pd_series.index)
```

### index를 활용한 데이터 탐색

```python
# loc(index 활용하여 access), iloc(integer location 활용하여 access)
print(pd_series.loc['a'])
print(pd_series.iloc[0])

print(pd_series.loc[['a','c']])
print(pd_series.iloc[[0,2]])
```

### 데이터 삭제

```python
pd_series.drop('b') # 임시 삭제
print(pd_series)
pd_series.drop('b', inplace=True) # 삭제
print(pd_series)
```

### Dataframe 만들기

```python
data = {
    'A': np.arange(15),
    'B': np.random.randint(low=0, high=15, size=(15)),
    'C': np.random.rand(15)
}

data_df = pd.DataFrame(data)
```

```python
data_df
```

```python
data_df.head(5) # default 5개
data_df.tail(5) # default 5개
data_df.shape
```

```python
# indexing
print(data_df[1:3])
print()
print(data_df.loc[1][['B', 'C']])
```

### 기타

```python
data_df.drop('D', axis=1, inplace=True) # 'D'라는 '열'을 inplace로 삭제하겠다
data_df.head()
```
